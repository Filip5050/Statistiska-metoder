```python
import numpy as np
import matplotlib.pyplot as plt

# Simulated Data: Advertising and Sales
tv_spend = np.array([0, 50, 100, 150, 200, 250, 300])  # TV ad spend in $1,000s
sales = np.array([7, 9, 12, 14, 17, 20, 24])  # Sales in $1,000s

# Step 1: Simple Linear Regression Coefficients
n = len(tv_spend)  # Number of observations
x_mean = np.mean(tv_spend)
y_mean = np.mean(sales)

# Calculating the slope (beta_1) and intercept (beta_0)
beta_1 = np.sum((tv_spend - x_mean) * (sales - y_mean)) / np.sum((tv_spend - x_mean) ** 2)
beta_0 = y_mean - beta_1 * x_mean

# Predicted values
sales_pred = beta_0 + beta_1 * tv_spend

# Step 2: Residuals
residuals = sales - sales_pred

# Step 3: Residual Sum of Squares (RSS)
rss = np.sum(residuals ** 2)

# Step 4: Plotting the Linear Fit with Residuals
plt.figure(figsize=(12, 6))
plt.scatter(tv_spend, sales, color='red', label='Actual Data')  # Actual data points
plt.plot(tv_spend, sales_pred, color='blue', label='Fitted Line')  # Fitted regression line

# Adding residuals as grey lines
for i in range(n):
    plt.plot([tv_spend[i], tv_spend[i]], [sales[i], sales_pred[i]], color='grey', linestyle='--')

plt.title('Linear Regression: Sales vs TV Advertising')
plt.xlabel('TV Advertising Spend ($1,000s)')
plt.ylabel('Sales ($1,000s)')
plt.legend()
plt.grid(True)
plt.show()

# Step 5: Residual Plot
plt.figure(figsize=(12, 6))
plt.scatter(tv_spend, residuals, color='purple')
plt.axhline(y=0, color='black', linestyle='--')  # Reference line at 0
plt.title('Residual Plot')
plt.xlabel('TV Advertising Spend ($1,000s)')
plt.ylabel('Residuals (Actual - Predicted)')
plt.grid(True)
plt.show()


```

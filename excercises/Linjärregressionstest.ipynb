```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r"C:\Code\Statistiska metoder\Datasets\Salary_dataset.csv")
```

```python
X = df["YearsExperience"].values
Y = df["Salary"]
```

```python

# Skapa designmatrisen (lägger till en kolumn med ettor för interceptet)
X_design = np.column_stack((np.ones(len(X)), X))
```

```python
# Utför linjär regression
beta = np.linalg.lstsq(X_design, Y, rcond=None)[0]
```

```python
# Prediktioner
y_pred = X_design.dot(beta)
```

```python
residuals = Y - y_pred
```

```python
#Räkna ut medelvärdet, variance, standardavikelsen
y_mean = np.mean(Y)
variance = np.var(residuals, ddof=1)
std = np.std(residuals, ddof=1)
```

```python
#Räkna ut SSE, SSR, SST
SST = np.sum((Y - y_mean) ** 2)
SSE = np.sum(residuals **2)
SSR = SST-SSE
```

```python
#Räkna ut R^2
R_squared = SSR / SST
```

```python
print(f"Intercept: {beta[0]}")
print(f"Lutning: {beta[1]}")
print(f"Varians: {variance}")
print(f"Standardavvikelse: {std}")
print(f"SST (Total Sum of Squares): {SST}")
print(f"SSE (Sum of Squared Errors): {SSE}")
print(f"SSR (Sum of Squares for Regression): {SSR}")
print(f"R^2: {R_squared}")
```

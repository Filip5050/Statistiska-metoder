```python
import numpy as np

# Exempeldata
y_observed = np.array([10, 20, 30])  # Observerade värden
y_predicted = np.array([9, 18, 29])  # Predikterade värden

# Beräkna medelvärdet av de observerade värdena
y_mean = np.mean(y_observed)

# Beräkna TSS (Total Sum of Squares)
tss = np.sum((y_observed - y_mean) ** 2)

# Beräkna RSS (Residual Sum of Squares)
rss = np.sum((y_observed - y_predicted) ** 2)

# Beräkna R^2
r2 = 1 - (rss / tss)

print("R^2:", r2)
```

```python
import matplotlib.pyplot as plt

# Example data
X = np.array([1, 2, 3, 4, 5])  # Independent variable
y = np.array([2, 4, 5, 4, 5])  # Dependent variable

# Step 1: Add a column of ones to X for the intercept term
X_b = np.c_[np.ones((X.shape[0], 1)), X]  # Add intercept (bias term)

# Step 2: Compute coefficients using the normal equation
beta = np.linalg.inv(X_b.T @ X_b) @ X_b.T @ y  # Normal equation
beta_0, beta_1 = beta  # Intercept and slope

# Step 3: Make predictions
y_pred = X_b @ beta


# Step 5: Visualize the results
plt.scatter(X, y, color="blue", label="Data points")
plt.plot(X, y_pred, color="red", label="Fitted line")
plt.xlabel("X")
plt.ylabel("y")
plt.title("Linear Regression with NumPy")
plt.legend()
plt.show()
```

```python
x = np.array([1,2,3]) #x värden
y = np.array([2,3,5]) #y värden

x_mean = np.mean(x)
y_mean = np.mean(y)

numerator = np.sum((x - x_mean) * (y - y_mean))
deomiantor = np.sum((x-x_mean) **2)

m = numerator / deomiantor
b = y_mean - m * x_mean

print(f"Lutning (m): {m}")
print(f"Intercept (b): {b}")

# 2. Räkna ut de predicerade y-värdena för linjen
y_pred = m * x + b

# 3. Visualisera punkterna och linjen
plt.scatter(x, y, color="blue", label="Data Points")
plt.plot(x, y_pred, color="red", label=f"Linje: y = {m:.2f}x + {b:.2f}")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.title("Linjär Regression")
plt.show()
```
